```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt # data visualization
import seaborn as sns # statistical data visualization
%matplotlib inline

import warnings

warnings.filterwarnings('ignore')
```

### Load Train and Test Data

```python
#Load Train Data
df_train = pd.read_csv('titanic_train.csv')
df_train.info()
```

```python
df_train.head()
```

```python
df_test = pd.read_csv('titanic_test.csv')
df_test.info()
```

```python
df_test.head()
```

## EDA on Training Data

```python
#check if there are any NA Value
sns.heatmap(df_train.isna().transpose())
```

```python
# Drop unused Column
df_train.drop(['Cabin'],axis=1,inplace=True)
df_train.head()
```

```python
### Remove entries with NA and recheck again
df_train.dropna(inplace=True)
sns.heatmap(df_train.isna().transpose())
```

```python
df_train.head()
```

```python
## Sex need to be encoded using Dummy Encoding..How about embarked..Let's check the list of possible values in the fields
embarked_unique = df_train['Embarked'].unique()
parch_unique = df_train['Parch'].unique()
class_unique = df_train['Pclass'].unique()

print(f'Embarked : {embarked_unique}')
print(f'Parch : {parch_unique}')
print(f'Pclass : {class_unique}')
```

```python
#Dummies Encoding

df_dummies = pd.get_dummies(df_train[['Pclass', 'Sex','Embarked']],prefix='dm')
df_train = pd.concat([df_train,df_dummies],axis=1)
df_train.head()
```

```python
df_train.info()
```

```python
#Load data to X & y
X = df_train.drop(['PassengerId', 'Survived', 'Pclass','Sex','Ticket','Embarked'],axis = 1)
y = df_train[['Survived']]
print(f'X train shape : {X.shape}')
print(f'y train shape : {y.shape}')
```

```python
df_dummies = pd.get_dummies(X['Parch'],prefix='dm')
X = pd.concat([X,df_dummies],axis=1)
X.head()
```

```python
X.drop(['Name','Parch'],axis=1,inplace=True)
X.head()
```

```python
#Split Data 

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
print(f'X train shape : {X_train.shape}')
print(f'y train shape : {y_train.shape}')
print(f'X test  shape : {X_test.shape}')
print(f'y test  shape : {y_test.shape}')
```

```python
#Let's Train it :)
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=0)
rfc.fit(X_train,y_train)
```

```python
#Get accuracy, precission, recall and F1 score
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score


y_predict = rfc.predict(X_test)
print (f'Accuracy: {accuracy_score(y_test,y_predict)}')
print (f'Precision: {precision_score(y_test,y_predict)}')
print (f'Recall: {recall_score(y_test,y_predict)}')
print (f'f1: {f1_score(y_test,y_predict)}')
```

```python
#Since we want to predict if the people survived or not, we aim for Higher Precission..Let's check how many trees it takes to get the accurate data.
```

```python
result = {'tree_num' : [], 'accuracy':[], 'precision':[], 'recall':[], 'f1':[]}
for n_est in range(50,500):
    rfc = RandomForestClassifier(n_estimators = n_est,random_state=0)
    rfc.fit(X_train,y_train)
    y_predict = rfc.predict(X_test)
    result['tree_num'].append(n_est)
    result['accuracy'].append(accuracy_score(y_test,y_predict))
    result['precision'].append(precision_score(y_test,y_predict))
    result['recall'].append(recall_score(y_test,y_predict))
    result['f1'].append(recall_score(y_test,y_predict))
    
n_est_res = pd.DataFrame(result)
n_est_res.head()
```

```python
fig,axes = plt.subplots(2,2, figsize=(18,10))
sns.lineplot(ax=axes[0,0],data=n_est_res,x='tree_num',y='accuracy')
sns.lineplot(ax=axes[0,1],data=n_est_res,x='tree_num',y='precision')
sns.lineplot(ax=axes[1,0],data=n_est_res,x='tree_num',y='recall')
sns.lineplot(ax=axes[1,1],data=n_est_res,x='tree_num',y='f1')
```

```python
#Looks like its between 150-200 Instances that having the highest precision.
```

```python
n_est_res1 = n_est_res.loc[n_est_res['tree_num'] >= 150]
n_est_res1 = n_est_res1.loc[n_est_res1['tree_num'] <= 200]
#plot it again with bar plot
fig,axes = plt.subplots(2,2, figsize=(18,10))
sns.lineplot(ax=axes[0,0],data=n_est_res1,x='tree_num',y='accuracy')
sns.lineplot(ax=axes[0,1],data=n_est_res1,x='tree_num',y='precision')
sns.lineplot(ax=axes[1,0],data=n_est_res1,x='tree_num',y='recall')
sns.lineplot(ax=axes[1,1],data=n_est_res1,x='tree_num',y='f1')
```

```python
#Narrow it down again (295-300) :)
```

```python
n_est_res1 = n_est_res.loc[n_est_res['tree_num'] >= 160]
n_est_res1 = n_est_res1.loc[n_est_res1['tree_num'] <= 170]
#plot it again with bar plot
fig,axes = plt.subplots(2,2, figsize=(18,10))
sns.lineplot(ax=axes[0,0],data=n_est_res1,x='tree_num',y='accuracy')
sns.lineplot(ax=axes[0,1],data=n_est_res1,x='tree_num',y='precision')
sns.lineplot(ax=axes[1,0],data=n_est_res1,x='tree_num',y='recall')
sns.lineplot(ax=axes[1,1],data=n_est_res1,x='tree_num',y='f1')
```

```python
# now we get our sweet spot (161)
```

```python
n_est_res.loc[n_est_res['tree_num'] >= 160].head()
```

```python
rfc = RandomForestClassifier(n_estimators = 161,random_state=0)
rfc.fit(X_train,y_train)
y_predict = rfc.predict(X_test)
    
print (f'Accuracy: {accuracy_score(y_test,y_predict)}')
print (f'Precision: {precision_score(y_test,y_predict)}')
print (f'Recall: {recall_score(y_test,y_predict)}')
print (f'f1: {f1_score(y_test,y_predict)}')   
```
